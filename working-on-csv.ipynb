```python
name=""
def rename(name):
    if name =="Royal Challengers Bangalore":
        return "RCB"
    elif name =="Kings XI Punjab":
        return "KXIP"
    elif name == "Delhi Daredevils":
        return "DD"
    elif name =="Mumbai Indians":
        return "MI"
    elif name == "Kolkata Knight Riders":
        return "KKR"
    elif name == "Delhi Capitals":
        return 'DC'
    elif name == "Rajasthan Royals":
        return 'RR'
    elif name == "Sunrisers Hyderabad":
        return "SRH"
    elif name == "Deccan Chargers":
        return 'DCH'
    elif name =="Chennai Super Kings":
        return 'CSK'
    else:
        return name
rename(name)
cols = ["team1", "team2", "toss_winner", "winner"]
data3=pd.read_csv("/kaggle/input/datasets/anubhavpathak0411/ipl-2008-2020/IPL Matches 2008-2020.csv",converters={col:rename for col in cols}
                 )
data3
```
